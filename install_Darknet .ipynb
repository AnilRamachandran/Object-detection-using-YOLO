```python
# check the installation by running
!./darknet

# Expected output: 
# usage: ./darknet <function>
```

```python
# run the detector 
# where the config file for YOLO is in the cfg/ subdirectory
!./darknet detect cfg/yolov3.cfg yolov3.weights data/dog.jpg
```

```python
# detected objects is saved in predictions.jpg file
%cd /anil/object_detection/darknet
%pylab inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
img=mpimg.imread('/anil/object_detection/darknet/predictions.jpg')
imgplot = plt.imshow(img)
plt.show()
```
